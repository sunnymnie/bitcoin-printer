```python
db = pd.read_csv("../data/BTCUSDT-dollar-bars.csv")
vb = pd.read_csv("../data/BTCUSDT-volume-bars.csv")
```

```python
tb = pd.read_csv("../data/BTCUSDT-minute-target-ui.csv")
```

```python
tb.head(5)
```

```python
len(tb)
```

```python
tb2 = tb[['timestamp', 'open', 'high', 'low', 'close', 'volume']][300000:300200]
```

```python
fig = go.Figure(data=[go.Candlestick(x = tb2['timestamp'],
                                     open = tb2['open'], 
                                     high = tb2['high'],
                                     low = tb2['low'], 
                                     close = tb2['close'])])
fig.update_layout(
    autosize=True,
    height=800,
)
fig.show()
```
