```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import useful_functions as uf
import classification_report_generator as crg
from ta.momentum import WilliamsRIndicator
```

```python
df = pd.read_csv("../data/BTCUSDT-hour-target.csv")
```

## Variations and default correlationsÂ¶

```python
def variations(windows:list[int]):
    """
    returns variations of atr indicator with given windows
    """
    name = "wr_"
    for w in windows:

        i = WilliamsRIndicator(df.high, df.low, df.close, w)
        sma_name = name + str(w)
        df[sma_name] = i.williams_r()
#         df[sma_name + "_log"] = list(map(lambda x: uf.log_abs(x*100, zeros=True), i.rsi()))
```

```python
variations([5, 10, 15, 20, 25, 30, 50, 100, 200])
# variations([50])
```

```python
indicators = list(df.columns)[13:]
indicators.append("target")
d = df[list(indicators)].copy()
corr_matrix = d.corr()
fig, ax = plt.subplots(figsize=(20, 15))
ax = sns.heatmap(corr_matrix,
                 annot=True,
                 linewidths=0.5,
                 fmt=".2f",
                 cmap="YlGnBu")
```

```python
fig,ax = plt.subplots(figsize=(20, 10))
ax.plot(df.wr_50,
        color="green")
ax.set_xlabel("time")
ax.set_ylabel("atr")

ax2=ax.twinx()
ax2.plot(df.close,
         color="gray")

ax2.set_ylabel("close")
# fig.legend(["a", "b", "c", "d"])
plt.show()
```
