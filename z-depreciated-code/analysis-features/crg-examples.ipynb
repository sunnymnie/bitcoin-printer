# Classification report generator examples
Example uses of classification_report_generator.py

```python
import classification_report_generator as crg
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/BTCUSDT-hour-target-classification.csv")
df.dropna(inplace=True)
X = df.drop(["target", "timestamp"], axis=1)
y = df["target"]
```

```python
crg.plot_conf_matrix(X, y)
```

```python
crg.generate_report(X, y, remind=True)
```

```python
crg.feature_importance(X, y, plot=False)
```

```python
crg.feature_importance(X, y, plot=True)
```

```python
crg.foward_test(X, y, X["close"], mistakes=False)
```

```python
crg.foward_test(X, y, X["close"], mistakes=True)
```
