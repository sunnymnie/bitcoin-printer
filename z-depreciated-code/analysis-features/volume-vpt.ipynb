# VolumePriceTrendIndicator

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import useful_functions as uf
import classification_report_generator as crg
from ta.volume import VolumePriceTrendIndicator
```

```python
df = pd.read_csv("../data/BTCUSDT-hour-target.csv")
```

## Variations and default correlations

```python
i = VolumePriceTrendIndicator(df.close, df.volume)
df["vpt"] = i.volume_price_trend()
```

```python
indicators = list(df.columns)[13:]
indicators.append("target")

d = df[list(indicators)].copy()
corr_matrix = d.corr()
fig, ax = plt.subplots(figsize=(10, 8))
ax = sns.heatmap(corr_matrix,
                 annot=True,
                 linewidths=0.5,
                 fmt=".2f",
                 cmap="YlGnBu")
```

```python
fig,ax = plt.subplots(figsize=(20, 10))
ax.plot(df.vpt_amplified,
        color="green")
ax.set_xlabel("time")
ax.set_ylabel("atr")

ax2=ax.twinx()
ax2.plot(df.close,
         color="gray")

ax2.set_ylabel("close")
# fig.legend(["a", "b", "c", "d"])
plt.show()
```

```python
df["vpt_amplified"] = list(map(lambda x: uf.log_abs(x*100, zeros=True), df.vpt))
```
