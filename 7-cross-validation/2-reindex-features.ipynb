# Reindex features

```python
import pandas as pd
import numpy as np
from IPython.display import clear_output
```

```python
index = pd.read_csv("../data/weighted_seq_bootstrap.csv", index_col=0, parse_dates=[0, 5]).index
ml = pd.read_csv("../data/metalabels.csv", index_col=0, parse_dates=True)
```

```python
indicators = pd.DataFrame(index=index, columns=[list(map(lambda x: "i"+str(x), [0]))])
i = 0
while True:
    try:
        print(f"working on indicator i{i}")
        indicators["i"+str(i)] = pd.read_csv(f"../data/i{i}.csv", index_col=0, parse_dates=True, squeeze=True).reindex(index)
        clear_output()
    except:
        print("finished!")
        break
    i += 1
```

```python
indicators["side"] = ml.side.reindex(index)
indicators["sc"] = ml.sc.reindex(index)
indicators["random"] = np.random.uniform(size=len(indicators))
```
